```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/Zomato-data-.csv")

print("Dataset Loaded Successfully")
```

```python
print("First 5 Rows:")
print(df.head())

print("\nShape:")
print(df.shape)

print("\nColumns:")
print(df.columns)
```

```python
print("Dataset Information:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
print("Duplicate Rows:")
print(df.duplicated().sum())
```

```python
# Clean rate column
df["rate"] = df["rate"].str.replace("/5", "", regex=False)
df["rate"] = pd.to_numeric(df["rate"], errors="coerce")

print(df["rate"].head())
print(df["rate"].dtype)
```

```python
# Clean cost column
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].astype(str)
df["approx_cost(for two people)"] = df["approx_cost(for two people)"].str.replace(",", "", regex=False)
df["approx_cost(for two people)"] = pd.to_numeric(df["approx_cost(for two people)"], errors="coerce")

print(df["approx_cost(for two people)"].head())
print(df["approx_cost(for two people)"].dtype)
```

```python
# Fill numeric null values with mean
df["rate"].fillna(df["rate"].mean(), inplace=True)
df["votes"].fillna(df["votes"].mean(), inplace=True)
df["approx_cost(for two people)"].fillna(df["approx_cost(for two people)"].mean(), inplace=True)

print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
# Save cleaned dataset
df.to_csv("cleaned_zomato_data.csv", index=False)

print("Cleaned dataset saved successfully")
```
