```python
import numpy as np
import pandas as pd
```

```python
student_data = np.array([
    [1, 85, 5, 90],
    [2, 78, 4, 85],
    [3, 92, 6, 95],
    [4, 60, 2, 70],
    [5, 75, 3, 80]
])

print("Student Data:")
print(student_data)
```

```python
print("Shape:", student_data.shape)
print("Dimensions:", student_data.ndim)
print("Data Type:", student_data.dtype)

print("\nMarks Column:")
print(student_data[:, 1])

print("\nAverage Marks:")
print(np.mean(student_data[:, 1]))

print("\nMaximum Marks:")
print(np.max(student_data[:, 1]))

print("\nMinimum Marks:")
print(np.min(student_data[:, 1]))
```

```python
high_math = df[df["math score"] > 80]

print("Students with Math Score > 80:")
print(high_math.head())

print("\nTotal Students with Math Score > 80:")
print(len(high_math))
```

```python
# Average math score by gender
avg_math_gender = df.groupby("gender")["math score"].mean()

print("Average Math Score by Gender:")
print(avg_math_gender)
```
